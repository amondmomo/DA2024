## 지도학습, 비지도 학습, 강화 학습

지도학습 : 정답이 있음
- 예) 고양이 사진, 토끼 사진 등..

비지도 학습 : 정답이 없음, 상관관계 학습
- 예) 구매 이력으로 소득수준 파악, 취향 판단 등

강화 학습 : 데이터 입력x, 인공지능 <-> 환경 상호작용 / 보상과 벌을 통해 학습 개선
- 예) 알파고

## 기초개념

- 퍼셉트론 :
-- **입력값**(x1,x2,..)과 **가중치**(w1,w2..), **편향**(b,bias)을 이용해 출력값을 내는 수학적 모델.
-- 출력값이 0과 1로 출력됨.
-- 퍼셉트론이 하나이면 **단층 인공 신경망** 퍼셉트론이 여러개 이면 **다층 인공 신경망**

- 활성화 함수 : 해당 뉴런의 출력을 다음 뉴런으로 넘길지 결정
- 시그모이드 함수 : 뉴런의 출력값을 0과 1 사이로 고정


=> 가중치 : 입력의 중요도

=> 편향 : 활성화의 경계가 원점으로부터 얼마나 이동할지 결정
